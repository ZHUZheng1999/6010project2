```python
df = pd.read_csv('./GKX_20201231.csv')
df.set_index('DATE', drop=True, inplace=True)
data_set = df.loc['19570101':'20170101']
```

```python
sic_data = data_set[['permno', 'sic2']]
```

```python
industry = data_set.sic2.unique()
industry = industry[~np.isnan(industry)]
```

```python
industry_var = pd.DataFrame(0, index=data_set.index, columns=industry)
from tqdm import trange
for i in trange(len(sic_data)):
    if np.isnan(sic_data.iloc[i, 1]):
        continue
    else:
        num = sic_data.iloc[i, 1]
        idx = industry.tolist().index(num)
        industry_var.iloc[i, idx] = 1
```
